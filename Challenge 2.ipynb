## Analysing by VOFR % / Volume Fill rate will be total quantity shipped / total quantity ordered 

```python
vofr = df1[['customer_id','order_qty','delivery_qty','In Full','On Time','On Time In Full']]
vofr['total_order_qty'] = vofr.groupby('customer_id')['order_qty'].transform('sum')
vofr['total_delivery_qty'] = vofr.groupby('customer_id')['delivery_qty'].transform('sum')
vofr = vofr.drop(columns=['order_qty','delivery_qty'])
vofr
```

```python
vofr['%vofr'] = vofr.groupby('customer_id').apply(
    lambda x: (x['total_delivery_qty'] / x['total_order_qty']) * 100
).reset_index(level=0, drop=True)
```
